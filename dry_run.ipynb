```python
df_balance_mean = df["balance"].mean()
df["balance"] = df["balance"].map(lambda x: x + df_balance_mean)
df["balance"]
```

```python
def add_balance_mean(row):
    balance_mean = df["balance"].mean()
    return row["balance"] + balance_mean

df["balance"] = df.apply(add_balance_mean, axis=1)
df.head()
```

```python
def change_education(row):
    if row["education"] == "primary":
        return "p"
    elif row["education"] == "secondary":
        return "s"
    else:
        return "t"

df["education"] = df.apply(change_education, axis=1)
```

```python
def subtract_mean(col):
    if col.name == "balance":
        return col - col.mean()
    else:
        return col

df = df.apply(subtract_mean, axis=0)
```
