```python
import pandas as pd
```

```python
df = pd.read_csv('bank.csv')
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
unknown_counts = (df == "unknown").sum()
unknown_counts
```

```python
df.drop(columns=['contact', 'poutcome'], inplace=True)
df.head()
```

```python
df.replace("unknown", pd.NA, inplace=True)
df.dropna(subset=['job', 'education'],inplace=True)
# optionally we could do this, insteading of dropping replace with the mode
# df["job"].fillna(df["job"].mode().iloc[0], inplace=True)
# df["education"].fillna(df["education"].mode().iloc[0], inplace=True)
```

```python
df["job"].value_counts()
df["job"].replace("admin.", "admin", inplace=True)
```

```python
age_bins = [0, 29, 60, 100]
age_labels = ['Under_30', '30_to_60', 'Above_60']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
```

```python
df['has_previous_contact'] = df['previous'] > 0
```

```python
df['last_contact_date'] = df['day'].astype(str) + '-' + df['month'].str.title()
```

```python
df['is_high_balance'] = df['balance'] > 1000
```

```python
#Call Efficiency Feature
df['efficiency'] = df['duration'] / df['campaign']
```

```python
#New Contact Flag
df['is_first_contact'] = df['pdays'] == -1
```

```python
df['edu_marital'] = df['education'] + '_' + df['marital']
```
